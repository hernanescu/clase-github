# Git: qué es y para qué sirve
- Git es un sistema gratuito y abierto de control de versiones.
- Todos los developers interactúan con esto de alguna u otra manera.

#### ¿Control de versión? ¿Con qué se come?
- Básicamente es una forma que tenemos para trackear nuestros cambios en el código.
- La idea es guardar versiones iniciales en Git, y a medida que vamos progresando podemos ir guardando distintas versiones. A medida que el código evoluciona, podemos ver todos los cambios que hicimos. 
- Esto es excesivamente importante para entender qué fue pasando, buscar versiones anteriores en caso de que algo falle, etc.


#### Algunos términos
- Directorio: Una carpeta.
- Terminal o Command Line: la interfaz para comandos de texto.
- CLI: Command Line Interface.
- Repositorio: Suele referirse a un repositorio de Git. Una carpeta donde está todo el proyecto.
- GitHub: Git es la herramienta con la que hacemos tod oesto, y GitHub es un sitio web donde hosteás tus repositorios online y te permite no sólo hacer un portfolio sino también trabajar de manera colaborativa (ya vamos a ver eso!).


#### Algunos comandos de Git
- clone: Traer un repositorio que está hosteado en otro lugar (como GitHub) a tu entorno local (mayormente tu computadora).
- add: Trackear archivos y cambios en Git.
- commit: Guardar los archivos modificados. "Commiteaste el cambio?".
- push: Subir los commits de git a un repositorio remoto, como GitHub. El opuesto de pull.
- pull: Bajar los cambios del repo remoto a la máquina local. El opuesto de push.
